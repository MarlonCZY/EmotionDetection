```python
# Before running this, remember to create training, testing folders and all subfolders manually

import csv

csvFile = open("ordered_labels.csv", "r")
reader = csv.reader(csvFile)

image_path = '../../images/'
catelog_path = '../../classifiedData/training/'

for item in reader:
    if reader.line_num == 1:
        continue
    print(item[0], item[1], item[2])
    cmd = 'cp '+image_path+str(item[1]) + " "+catelog_path+str(item[2])
    print(cmd)
    os.system(cmd)
```
